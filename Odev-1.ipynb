```python
### tas-kagit-makas###

import random
cpu_skor = 0
oyuncu_skor = 0
mover= ['tas', 'kagit', 'makas']
cpu = mover[random.randint(0,2)]
while oyuncu_skor <= 9 and cpu_skor <= 9:

    oyuncu = input('Hamleni yap!; ')
    if cpu == oyuncu:
        print('Berabere')
    elif oyuncu == 'tas':
        if cpu == 'kagit':
            cpu_skor += 1 
            print('Cpu kazandi!')
               
        else:
            oyuncu_skor += 1
            print('sen kazandin!')
            
    elif oyuncu == 'kagit':
        if cpu == 'makas':
            cpu_skor += 1
            print('Cpu kazandi!')
            
        else:
            oyuncu_skor += 1
            print('sen kazandin!')
            
    elif oyuncu == 'makas':
        if cpu == 'tas':
            cpu_skor += 1
            print('cpu kazandi!')
            
        else:
            oyuncu_skor += 1
            print('sen kazandin!')
            

    
print('Sen:',oyuncu_skor, 'Cpu: ', cpu_skor)
    
    
```

```python

### Ders Notu ###k

ad = input('Lufen adinzi giriniz; ')
soyad = input('Lufen soyadinzi giriniz; ')
no = input('Lufen numaranizi giriniz; ')

mat_vize = int(input('Lufen mat vize notunu giriniz; '))
fen_vize = int(input('Lufen fen vize notunu giriniz; '))
bio_vize = int(input('Lufen bio vize giriniz; '))
fizik_vize = int(input('Lufen fizik vize notunu giriniz; '))

mat_fin = int(input('Lufen mat final notunu giriniz; '))
fen_fin = int(input('Lufen fen final notunu giriniz; '))
bio_fin = int(input('Lufen biyoloji final giriniz; '))
fizik_fin = int(input('Lufen fizik final notunu giriniz; '))




mat_son = round(mat_vize * 0.4 + mat_fin * 0.6, 2)
fen_son = round(fen_vize * 0.4 + fen_fin * 0.6 ,2)
bio_son = round(bio_vize * 0.4 + bio_fin * 0.6,2) 
fizik_son = round(fizik_vize * 0.4 + fizik_fin * 0.6,2)
 
print(f'Merhaba {ad} {soyad} ({no})')
if mat_son > 50:
    print(f'{mat_son} ile matematik dersinden gectiniz')
else:
    print(f'{mat_son} ile matematik dersinden kaldiniz')
    
if fen_son > 50:
    print(f'{fen_son} ile fen dersinden gectiniz')
else:
    print(f'{fen_son} ile fen dersinden kaldiniz')
    
if bio_son > 50:
    print(f'{bio_son} ile biyoloji dersinden gectiniz')
else:
    print(f'{bio_son} ile bijoloji dersinden kaldiniz')
          
if fizik_son > 50:
    print(f'{fizik_son} ile fizik dersinden gectiniz')
else:
    print(f'{fizik_son} ile fizik dersinden kaldiniz')
          
```
